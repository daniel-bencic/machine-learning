```python
%matplotlib inline
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
import numpy as np
import skimage.io as sio
```

### a.

```python
digits = datasets.load_digits()
digits.data.shape
```

1797 Pictures with 8x8 pixels.

```python
digits.target.shape
```

One label for each picture.

```python
digits.target_names
```

Every picture is labeled 0-9.

```python
for i in range(10):
    idx = np.random.randint(0, digits.images.shape[0])
    print("Label: " + str(digits.target[idx]))
    sio.imshow(digits.images[idx])
    sio.show()
```

### b.

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target)
clf = svm.SVC(gamma=0.015, C=1.0)
clf.fit(X_train, y_train)
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

The problem here is overfitting, because the training set gets classified 100% correct but the test set has only a accuracy of 37.78%.

```python
clf2 = svm.SVC(gamma=0.001, C=100)
clf2.fit(X_train, y_train)
clf2.score(X_train, y_train)
```

```python
clf2.score(X_test, y_test)
```

With this parameters the classifier classified both test and trainings set 100% correct, which seems like a good fit.

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(digits.data, digits.target)
clf = svm.SVC(gamma=0.015, C=1.0)
clf.fit(X_train2, y_train2)
clf.score(X_train2, y_train2)
```

```python
clf.score(X_test2, y_test2)
```

```python
clf2 = svm.SVC(gamma=0.001, C=100)
clf2.fit(X_train2, y_train2)
clf2.score(X_train2, y_train2)
```

```python
clf2.score(X_test2, y_test2)
```

The accuracy of the prediction varies slighlty between the different random splits.
